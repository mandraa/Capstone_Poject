```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd

# List untuk menyimpan DataFrame
dfs = {}

# Membaca setiap file yang diunggah dan menyimpannya dalam dictionary
for filename in uploaded.keys():
    df = pd.read_csv(filename)
    # Menggunakan nama file tanpa ekstensi sebagai kunci
    key = filename.rsplit('.', 1)[0]
    dfs[key] = df

# Kolom pusat untuk penggabungan
order_id_key = 'order_id'
customer_id_key = 'customer_id'
product_id_key = 'product_id'
seller_id_key = 'seller_id'
product_category_name_key = 'product_category_name'

# Menggabungkan dataset berdasarkan kunci yang sesuai
# Mulai dengan olist_orders_dataset
merged_df = dfs['olist_orders_dataset']

# Gabungkan dengan olist_order_items_dataset berdasarkan 'order_id'
if 'olist_order_items_dataset' in dfs:
    merged_df = pd.merge(merged_df, dfs['olist_order_items_dataset'], on=order_id_key, how='outer')

# Gabungkan dengan olist_order_payments_dataset berdasarkan 'order_id'
if 'olist_order_payments_dataset' in dfs:
    merged_df = pd.merge(merged_df, dfs['olist_order_payments_dataset'], on=order_id_key, how='outer')

# Gabungkan dengan olist_customers_dataset berdasarkan 'customer_id'
if 'olist_customers_dataset' in dfs:
    merged_df = pd.merge(merged_df, dfs['olist_customers_dataset'], on=customer_id_key, how='outer')

# Gabungkan dengan olist_products_dataset berdasarkan 'product_id'
if 'olist_products_dataset' in dfs:
    merged_df = pd.merge(merged_df, dfs['olist_products_dataset'], on=product_id_key, how='outer')

# Gabungkan dengan olist_sellers_dataset berdasarkan 'seller_id'
if 'olist_sellers_dataset' in dfs:
    merged_df = pd.merge(merged_df, dfs['olist_sellers_dataset'], on=seller_id_key, how='outer')

# Gabungkan dengan product_category_name_translation berdasarkan 'product_category_name'
if 'product_category_name_translation' in dfs:
    merged_df = pd.merge(merged_df, dfs['product_category_name_translation'], on=product_category_name_key, how='outer')

# Menampilkan beberapa baris dari DataFrame gabungan
print(merged_df.head())
```

```python
df = pd.DataFrame(merged_df)
df
```

```python
df.isnull().sum()
```

```python
# Kolom yang harus diisi dengan nilai dari 'order_purchase_timestamp'
columns_to_fill = [
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_approved_at'
]

# Isi kolom tertentu dengan nilai dari 'order_purchase_timestamp'
df[columns_to_fill] = df.apply(lambda row: row['order_purchase_timestamp']
                               if pd.isna(row[columns_to_fill]).all()
                               else row[columns_to_fill], axis=1)

# Hapus baris dengan 'order_purchase_timestamp' adalah 15
df = df[df['order_purchase_timestamp'] != 15]

df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Menyimpan DataFrame gabungan ke dalam satu file CSV
df.to_csv('Brazil_E-commerce.csv', index=False)

# Mengunduh file CSV gabungan
from google.colab import files
files.download('Brazil_E-commerce.csv')
```

```python
# Konversi kolom ke tipe data datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'], format='%Y-%m-%d %H:%M:%S')
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'], format='%Y-%m-%d %H:%M:%S')
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'], format='%Y-%m-%d %H:%M:%S')
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'], format='%Y-%m-%d %H:%M:%S')
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'], format='%Y-%m-%d %H:%M:%S')
df['shipping_limit_date'] = pd.to_datetime(df['shipping_limit_date'], format='%Y-%m-%d %H:%M:%S')
```
